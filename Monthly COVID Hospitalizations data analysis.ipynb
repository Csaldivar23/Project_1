```python
from pathlib import Path
import pandas as pd

monthly_covid_hos_df = pd.read_csv("..\Project_1\Resources\Monthly_Rates_of_Laboratory-Confirmed_COVID-19_Hospitalizations_from_the_COVID-NET_Surveillance_System_20240723.csv")
monthly_covid_hos_df
```

```python
unique = monthly_covid_hos_df["State"].unique()
unique
```

```python
#limiting data to states listed
states_to_include = ["Utah", "Tennessee", "Oregon", "Ohio", "New York", "New Mexico", "Minnesota", "Michigan", "Maryland", "Iowa", "Georgia", "Connecticut", "Colorado", "California"]
filtered_df = monthly_covid_hos_df[monthly_covid_hos_df['State'].isin(states_to_include)]
filtered_df
```

```python
#filter the columns to State, Season, WeekendDate, and WeeklyRate
filtered_df = filtered_df[['State', 'Season', '_YearMonth', 'MonthlyRate']]
filtered_df
```

```python
filtered_df['_YearMonth'] = pd.to_datetime(filtered_df['_YearMonth'].astype(str), format='%Y%m')
```
